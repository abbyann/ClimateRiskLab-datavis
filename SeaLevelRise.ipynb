# Visualizing sea level rise from NOAA’s Laboratory for Satellite Altimetry dataset (https://www.star.nesdis.noaa.gov/socd/lsa/SeaLevelRise/LSA_SLR_timeseries.php)

```python
# 12/02/2023 - Visualizing Data from NOAA's satellite dataset
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('slr_sla_gbl_keep_txj1j2_90.csv', header=5)
df2 = pd.read_csv('slr_sla_gbl_keep_txj1j2_90.csv', header=5)
df
```

```python
# remove leading decimals from "year" column cells
df['year'] = df['year'].apply(np.floor)
#int(df['year'])
df
```

```python
# plot sea level deviations for each satellite

x = df['year']
y = df[['TOPEX/Poseidon','Jason-1','Jason-2','Jason-3']]
plt.figure(figsize=(20, 20))
plt.plot(x, y, 'o', label=["TOPEX/Poseidon", "Jason-1", "Jason-2", "Jason-3"])

plt.title("Satellite-measured global mean sea level", fontsize=35)
plt.xlabel("Year", fontsize=30)
plt.ylabel("Sea level (mm)", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

The above graph shows each of the raw sea level deviation data points across time. It is cluttered since there are several points taken at each year, since I rounded all the x-points down to the year.

```python
# plot sea level deviations for each satellite

x = df2['year']
y = df2[['TOPEX/Poseidon','Jason-1','Jason-2','Jason-3']]
plt.figure(figsize=(20, 15))
plt.plot(x, y, 'o', label=["TOPEX/Poseidon", "Jason-1", "Jason-2", "Jason-3"])

plt.title("Satellite-measured global mean sea level", fontsize=35)
plt.xlabel("Year", fontsize=30)
plt.ylabel("Sea level (mm)", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

The scatterplot above shows each sea level deviation data point from the year 1992 to 2022. Each color represents a different satellite used to collect the data, as NASA engineered better versions of their technology over the years.

Caption: This scatterplot shows the global mean sea level measurements collected by NASA’s satellites TOPEX/Poseidon and the Jason series. There is an evident upward trend in these measurements, with sea levels having risen about 100 mm since 1992.
