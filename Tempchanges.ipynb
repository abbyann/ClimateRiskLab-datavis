# Visualizing surface temperature changes from Nasa's GISTEMP dataset (https://data.giss.nasa.gov/gistemp/)

```python
# 12/02/2023 - Visualizing Data from GISTEMP
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('GLB.Ts+dSST.csv', na_values = '***', skiprows=1, header=None, sep = ',', skipinitialspace = True)

new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
#df.set_index('Year', inplace=True)

df.columns
df['Year']
```

```python
# len(df) = 144
# python indentation is non-inclusive at end

# define empty vector
array_avg = []

# calculate average of all monthly temperature values Jan-Dec for each year 1880-2023
for i in range (len(df)):
  new_df = pd.DataFrame(df.iloc[i, 1:13])
  # axis condition defines axis or axes along which the means are computed. The default is to compute the mean of the flattened array.
  # .values method pulls out first value
  average = new_df.mean(axis=1).values[0]
  # append average temperature value to empty array_avg array
  array_avg.append(average)

# array_avg is a vector containing the average temperature extremes across each month within each year from 1880-2023
array_avg
```

```python
# make plot of average surface temperature over each year
# define x as each "Year" value
x = df['Year']
print(x)
print(array_avg)
plt.figure(figsize=(20, 20))
plt.plot(x, array_avg, 'bo')

plt.title("Average yearly temperature extremes for 1880-2023", fontsize=35)
plt.xlabel("Year", fontsize=30)
plt.ylabel("Averaged temperature deviations (\u2103)", fontsize=30)
plt.xticks(x[::10], fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

This plot shows the averaged temperature deviations for each year from 1880 to 2023. In the original dataset, a temperature deviation from a long-term global average was determined for each month within each year. In this plot, the average of these deviations for each month was calculated for each year, for simplicity. Therefore, this plot shows the mean yearly temperature deviations from a long-term global average temperature.

Caption: This scatterplot shows the mean yearly surface temperature
deviations for the years 1880 through 2023, in degrees Celcius. There is a clear upward trend in deviations from a long-term global average (defined by GISTEMP as a baseline value).
